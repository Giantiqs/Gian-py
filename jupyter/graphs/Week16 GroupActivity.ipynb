```python
# Week16 GroupActivity
# Cabillan, Melchor Rich
# Fontanoza, Ivann Allen
# Gargar, Gabriel
# Jalac, Joshua
# Tiqui, Michael Gian M.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel("FANG_earnings.xlsx", sheet_name = "FANG (2021 - 2022)")

plt.plot(df["Month"], df["Facebook"], label="Facebook", color="blue", marker="o")
plt.plot(df["Month"], df["Amazon"], label="Amazon", color="orange", marker="*")
plt.plot(df["Month"], df["Netflix"], label="Netflix", color="black", marker="x")
plt.plot(df["Month"], df["Google"], label="Google", color="violet", marker="*")

plt.xlabel("Month")
plt.ylabel("Earnings (USD)")
plt.title("FANG Earnings 2021 - 2022", fontsize=20)
plt.rcParams.update({'font.size': 8.5})
plt.rcParams['figure.figsize'] = [20, 7]
plt.legend()

plt.show()

df2 = pd.read_excel("FANG_earnings.xlsx", sheet_name = "FANG (yearly)")

plt.plot(df2["Year"], df2["Facebook"], label="Facebook", color="blue", marker="o")
plt.plot(df2["Year"], df2["Amazon"], label="Amazon", color="orange", marker="*")
plt.plot(df2["Year"], df2["Netflix"], label="Netflix", color="black", marker="x")
plt.plot(df2["Year"], df2["Google"], label="Google", color="violet", marker="*")

plt.xlabel("Year")
plt.ylabel("Earnings (USD)")
plt.title("FANG Earnings 2016 - 2022", fontsize=20)
plt.rcParams.update({'font.size': 8.5})
plt.rcParams['figure.figsize'] = [20, 7]
plt.legend()

plt.show()
```
