```python
import matplotlib.pyplot as pp
import numpy as np

year = np.array(["1st year", "2nd year", "3rd year", "4th year"])
students = np.array([1203, 525, 659, 562])

colors = ["red", "blue", "yellow", "brown"]

pp.bar(year, students, color=colors)
pp.title("Number of students in CITCS", fontsize=10)
pp.xlabel("Year level", fontsize=10)
pp.ylabel("Students", fontsize=10)
pp.grid(True)
pp.show()
```

```python
import matplotlib.pyplot as pp
import numpy as np

year = np.array(["1st year", "2nd year", "3rd year", "4th year"])
students = np.array([1203, 525, 659, 562])

pp.plot(year, students, color="red", marker="o")
pp.title("Number of students in CITCS", fontsize=10)
pp.xlabel("Year level", fontsize=10)
pp.ylabel("Students", fontsize=10)
pp.grid(True)
pp.show()
```

```python
import matplotlib.pyplot as pp
import numpy as np

year = np.array(["1st year", "2nd year", "3rd year", "4th year"])
students = np.array([1203, 525, 659, 562])

colors = ["red", "blue", "yellow", "brown"]

pp.pie(students, autopct="%1.1f%%", colors=colors)
pp.title("Number of students")
pp.show()
```
